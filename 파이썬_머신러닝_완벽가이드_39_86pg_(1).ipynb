[개념정리]
===========

DataFrame : 여러 개의 행과 열로 이뤄진 2차원 데이터를 담는 데이터 구조체

read_csv()
  * 필드 구분 문자가 콤마  (sep="~~")
  * 호출 시 파일명 인자로 들어온 파일을 로딩해서 dataframe 객체로 반환

read_table() : 필드 구분 문자가 탭

read_fwf() : 고정 길이 기반의 칼럼 포맷을 dataframe으로 로딩하기 위한 API

info()메서드 : 총 데이터 건수, 데이터 타입, null 건수

describe()메서드 : 칼럼별 숫자형 데이터값의 n-percentile 분포도, 평균값, 최대값, 최소값

value_counts() : 지정된 칼럼의 데이터값 건수를 반환

인덱스
- 모든 series와 dataframe은 반드시 인덱스를 가짐
- 인덱스는 dataframe, series가 만들어진 후에도 변경 가능
- 고유성이 보장돼야 함

dataframe은 칼럼명 있음

딕셔너리를 dataframe으로 변환할 때)
- key는 문자열 칼럼명으로 매핑, value는 리스트 형(또는 ndarray) 칼럼 데이터로 매핑
- key값은 칼럼명, 각 value는 각 칼럼 데이터로 매핑됨

dataframe을 리스트로 변환 -> tolist()

dataframe을 딕셔너리로 변환 -> to_dict()

dataframe에서 데이터의 삭제 : drop() 메서드

- DataFrame.drop(labels=None, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise')
- axis 0 : 로우방향 축 / axis 1: 칼럼방향 축
- index : drop할 특정 로우, 칼럼 지정
- inplace=False이므로 다시 원본을 출력하면 drop되지 않은 상태가 출력됨

index 객체
- index는 단일값 반환, 슬라이싱 가능
- index 객체는 함부로 변경 불가
- index는 연산에서 제외 / 오직 식별용

reset_index()메서드 : 새롭게 인덱스를 연속 숫자형으로 할당

dataframe 뒤에 있는 []는 칼럼만 지정할 수 있는 '칼럼 지정 연산자'

dataframe의 인덱스값 : 명칭 기반 인덱싱

iloc : 위치 기반 인덱싱 -> 행,열로 정수, 정수형 슬라이싱, 팬시 리스트 값 가능 (불린 인덱싱 지원 x)

loc : 명령 기반 인덱싱 -> 행에는 dataframe index값 , 열에는 칼럼 명
- 슬라이싱하면 -> 시작~종료포함

and &, or |, not ~

sort_values() : dataframe과 series의 정렬

결손 데이터 : NULL (넘파이에서는 NaN)
- insa() : NaN 여부 확인
- fillna() : NaN값을 다른 값으로 대체

다른 aggregation 함수를 적용하려면 agg() 이용

lambda 함수 : 함수의 선언과 함수 내의 처리를 한 줄의 식으로 간단하게 나타내기 가능

map() 함수 : lambda 식을 이용할 때 여러 개의 값을 입력 인자로 사용해야 할 경우 이용

```python
# pg.82
def get_square(a):
  return a**2

print('3의 제곱은:', get_square(3))
```

```python
lambda_square = lambda x : x**2   # lambda 함수 : 함수의 선언과 함수 내의 처리를 한 줄의 식으로 간단하게 나타내기 가능
print('3의 제곱은:', lambda_square(3))
```

```python
a = [1,2,3]
squares = map(lambda x : x**2, a)   # map() 함수 : lambda 식을 이용할 때 여러 개의 값을 입력 인자로 사용해야 할 경우 이용
list(squares)
```

```python
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x : len(x))
titanic_df[['Name','Name_len']].head(3)
```

```python
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x <= 15 else 'Adult')
titanic_df[['Age', 'Child_Adult']].head(8)
```

```python
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : 'Child' if x<=15 else ('Adult' if x<=60 else 'Elderly'))
titanic_df['Age_cat'].value_counts()
```

```python
# 나이에 따라 세분화된 분류를 수행하는 함수 생성
def get_category(age):
  cat = ''
  if age <= 5 : cat = 'Baby'
  elif age <= 12 : cat = 'Child'
  elif age <= 18 : cat = 'Teenager'
  elif age <= 25 : cat = 'Student'
  elif age <= 35 : cat = 'Young Adult'
  elif age <= 60 : cat = 'Adult'
  else : cat = 'Elderly'

  return cat

# lambda 식에 위에서 생성한 get_category() 함수를 반환값으로 지정
# get_categort(X)는 입력값으로 'Age' 칼럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age','Age_cat']].head()
```
