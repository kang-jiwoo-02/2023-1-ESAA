앤스콤 데이터 집합 : 4개의 데이터 그룹으로 구성 + 차이 구분 위해서는 그래프의 시각화가 필요함!

앤스콤 데이터 집합은 seaborn 라이브러리에 포함되어 있음

load_dataset('a') : a 문자열을 전달하여 앤스콤 데이터 집합을 불러옴

matplotlib 라이브러리 : 그래프 그리기 가능

plot 메서드 : 선 그래프 (o : 점 그래프 가능)

fig = plt.figure()    # 그래프 격자가 위치할 기본 틀 만들기
plt.subplots() : 기본 틀 만들기

add_subplot 메서드
- add_subplot 메서드로 그래프 격자 그려줌
- add_subplot(기본 틀의 행 크기, 기본 틀의 열 크기, 기본 틀에서의 위치)

set_title 메서드 : 그래프 이름 전달해줌

suptitle 메서드 : 기본 틀에 제목 추가해줌

tight_layout 메서드 : 각 그래프의 레이아웃 조절가능

set_xlabel('a')  /  set_ylabel('b')

set_axis_labels() : x, y축 이름 한꺼번에 지정가능

히스토그램
- 데이터프레임의 열 데이터 분포와 빈도를 살펴보는 용도로 자주 사용하는 그래프

- 일변량 그래프 : 변수가 하나인 그래프 (x축 : frequency / y축 : 변수)
  - 자료명.hist(자료명['칼럼명'], bins=n) -> x축 간격 : n
- 이변량 그래프 : 변수가 두개인 그래프
  - 산점도 그래프 : scatter 메서드에 열들 전달해서 그래프 그림
  - scatter()
  - 박스 그래프 : 이산형 변수(명확하게 구분 가능)와 연속형 변수(명확하게 셀 수 없는 범위의 값)를 함께 사용하는 그래프

- 다변량 그래프 : 3개 이상의 변수를 사용한 그래프

산점도 그래프의 색상 지정하는 값으로 문자열은 사용 불가

distplot() : 히스토그램 + 밀집도 그래프 그려줌
- kde=False : 밀집도 그래프를 제외하고 히스토그램만 그려줌
- hist=False : 히스토그램을 제외하고 밀집도 그래프만 그려줌
- rug=True : 양탄자 그래프를 그려줌
  - 양탄자 그래프 : 그래프의 축에 동일한 길이의 직선을 붙여 데이터의 밀집 정도를 표현한 그래프

countplot() : count그래프 그려줌

regplot() : 산점도그래프 + 회귀선 그려줌
- fit_reg=False : 회귀선 제거해줌

joinplot() : 산점도그래프 + 히스토그램 그려줌

육각 그래프
- 2차원 표면에 육각형으로 데이터를 쌓아 표현하는 그래프
- 특정 데이터의 개수가 많아지면 점점 진한 색으로 표현
- jointplot에 kind="hex"
- hexbin()
- gridsize : 육각 그래프의 육각형 크기 조절

kdeplot() : 이차원 밀집도 그려줌
* shade=True : 그래프에 음영효과 그려줌

barplot() : 바 그래프 그려줌

boxplot() / box() : 박스 그래프 그려줌

바이올린 그래프
- 박스 그래프에 커널 밀도를 추정함
- 박스 그래프와 달리, 데이터 분산이 모호하게 표현되지 x
- violinplot()

관계 그래프
- pairplot()
- 대각선을 기준으로 절반이 중복됨
- map_upper()  /  map_lower()  /  map_diag()

hue=색상 사용할 열 이름 : 그래프에 색 추가

scatter_kws=딕셔너리 형태의 인잣값 : 그래프의 크기, 모양 조절

markers=[표현하고자 하는 기호]

그룹별로 그래프 나누려면 : col, col_warp 인자 설정하기
- col : 데이터 그룹을 구분할 열 지정
- col_wrap : 그래프를 그릴 열의 최댓값 지정

FacetGrid(데이터프레임명, 그룹 구분할 열 이름)
-> 
객체.map(그래프 종류, x축으로 지정할 열 이름)

alpha : 투명도  /  bins : x축의 간격

set_style() : 그래프의 스타일 변경 가능 / 한 번만 실행하면 계속 그렇게 유지됨

```python

# pg.88
# seaborn 라이브러리의 tips 데이터 집합 이용해서 ~

tips = sns.load_dataset("tips")   # tips 데이터 불러오기
print(tips.head())
print(type(tips))
```

```python
fig = plt.figure( )
axes1 = fig.add_subplot(1,1,1)
```

```python
axes1.hist(tips['total_bill'], bins=10)   # 자료명.hist(자료명['칼럼명'], bins=n) -> x축 간격 : n
axes1.set_title('Histogram of Total Bill') 
axes1.set_xlabel('Frequency') 
axes1.set_ylabel('Total Bill')

fig
```

```python
scatter_plot = plt.figure( )
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'], tips['tip'])    # 산점도 그래프 만드는 법 : scatter 메서드에 열 전달
axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

```python
# 박스 그래프 : 이산형 변수(명확하게 구분 가능)와 연속형 변수(명확하게 셀 수 없는 범위의 값)를 함께 사용하는 그래프

boxplot = plt.figure( )
axesl = boxplot.add_subplot(1,1,1)
axesl.boxplot([tips[tips['sex'] == 'Female']['tip'],
               tips[tips['sex'] == 'Male']['tip']],
               labels = ['Female','Male'])

axesl.set_xlabel('Sex') 
axesl.set_ylabel('Tip') 
axesl.set_title('Boxplot of Tips by Sex')
```

```python
# pg.92
# 다변량 그래프 그리기

def recode_sex(sex) :   # 산점도 그래프의 색상 지정하는 값으로 문자열은 사용 불가
  if sex == 'Female' :
    return 0
  else :
    return 1
```

```python
tips['sex_color'] = tips['sex'].apply(recode_sex)
```

```python
scatter_plot = plt.figure( )
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
  x=tips['total_bill'],
  y=tips['tip'],
  s=tips['size'] * 10,    # s : 점의 크기
  c=tips['sex_color'],    # c : 점의 색상
  alpha=0.5)              # alpha : 점의 투명도

axesl.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axesl.set_xlabel('Total Bill')
axesl.set_ylabel('Tip')
```

```python
# pg.95
# 단변량 그래프 그리기 - 히스토그램
import seaborn as sns
tips = sns.load_dataset("tips")
```

```python
ax = plt.subplots( )    # subplots() : 기본틀 만들어줌
ax = sns.distplot(tips['total_bill'])   # displot() : 히스토그램 그려줌
ax.set_title('Total Bill Histogram with Density Plot')
```

```python
ax = plt.subplots( )
ax = sns.distplot(tips['total_bill'], kde=False)    # kde=False : 밀집도 그래프를 제외하고 히스토그램만 그려줌
ax.set_title('Total Bill Histogram') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Frequency')
```

```python
ax = plt.subplots( )
ax = sns.distplot(tips['total_bill'], hist=False)   # hist=False : 히스토그램을 제외하고 밀집도 그래프만 그려줌 
ax.set_title('Total Bill Density') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Unit Probability')
```

```python
ax = plt.subplots( )
ax = sns.distplot(tips['total_bill'], rug=True)   # rug=True : 양탄자 그래프를 그려줌
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')# (양탄자 그래프 : 그래프의 축에 동일한 길이의 직선을 붙여 데이터의 밀집 정도를 표현한 그래프)
```

```python
ax = plt.subplots( )
ax = sns.countplot('day', data=tips)    # countplot() : count그래프 그려줌
ax.set_title('Count of days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```

```python
# pg.99
# 이변량 그래프 그리기
ax = plt.subplots( )
ax = sns.regplot(x='total_bill', y='tip', data=tips)    # regplot() : 산점도그래프 + 회귀선 그려줌
ax.set_title('Scatterplot of Total Bill and Tip') 
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax = plt.subplots( )
ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False)   # fit_reg=False : 회귀선 제거해줌
ax.set_title('Scatterplot of Total Bill and Tip') 
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
joint = sns.jointplot(x='total_bill', y='tip', data=tips)   # joinplot() : 산점도그래프 + 히스토그램 그려줌
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')    # set_axis_labels() : x, y축 이름 한꺼번에 지정가능
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
# 육각 그래프 : 2차원 표면에 육각형으로 데이터를 쌓아 표현하는 그래프
# : 특정 데이터의 개수가 많아지면 점점 진한 색으로 표현

hexbin = sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex") # jointplot에 kind="hex": 육각 그래프 그려줌
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
# 이차원 밀집도
ax = plt.subplots()
ax = sns.kdeplot(data=tips['total_bill'],data2=tips['tip'],shade=True)    # kdeplot() : 이차원 밀집도 그려줌
ax.set_title('Kernal Density Plot of Total Bill and Tip')                 # shade=True : 그래프에 음영 효과
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
# 바 그래프 -> 지정한 변수의 평균을 계산해서 그릴 수 있음
ax = plt.subplots( )
ax = sns.barplot(x='time', y='total_bill', data=tips)   # barplot() : 바 그래프 그려줌
ax.set_title('Bar plot of average total bill for time of day') 
ax.set_xlabel('Time of day') 
ax.set_ylabel('Average total bill')
```

```python
# 박스 그래프
ax = plt.subplots( )
ax = sns.boxplot(x='time', y='total_bill', data=tips)   # boxplot() : 박스 그래프 그려줌
ax.set_title('Boxplot of total bill by time of day') 
ax.set_xlabel('Time of day') 
ax.set_ylabel('Total Bill')
```

```python
ax = plt.subplots( )
ax = sns.violinplot(x='time', y='total_bill', data=tips)    # 바이올린 그래프 : 박스 그래프에 커널 밀도를 추정함 -> 데이터 분산이 모호하게 표현되지 x
ax.set_title('Violin plot of total bill by time of day')    # violinplot() : 바이올린 그래프 그려줌
ax.set_xlabel('Time of day') 
ax.set_ylabel('Total Bill')
```

```python
# 관계 그래프 : 대각선을 기준으로 절반이 중복됨
fig = sns.pairplot(tips)    # pairplot() : 관계 그래프 그려줌
```

```python
pair_grid = sns.PairGrid(tips) 
pair_grid = pair_grid.map_upper(sns.regplot)    # map_upper() : 대각선을 기준으로 위 쪽에 그릴 그래프 지정
pair_grid = pair_grid.map_lower(sns.kdeplot)    # map_lower() : 대각선을 기준으로 아래 쪽에 그릴 그래프 지정
pair_grid = pair_grid.map_diag(sns.distplot, rug=True)    # map_diag() : 대각선에 그릴 그래프 지정
plt.show()
```

```python
fig = plt.figure( )

seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']
for idx, style in enumerate(seaborn_styles):
  plot_position = idx + 1
  with sns.axes_style(style):
    ax = fig.add_subplot(2, 3, plot_position)
    violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
    violin.set_title(style)

fig.tight_layout()
```
